### Guessing a letter A-Z  
**check_guess()** takes 2 string arguments: **letter and guess** (both expect single alphabetical character)   
    - if guess is not an alpha character print invalid and return False
    - test and print if guess is "high" or "low" and return False
    - test and print if guess is "correct" and return True

```python
# [ ] create check_guess()
def check_guess(letter,guess):
    guess = str(guess)
    if guess.isalpha():
        #guess is alphabetical letters
        if len(guess)==1:
            #guess is one character
            #TODO check the guess
            if guess == letter:
                print("Correct")
                return True
            elif guess > letter:
                print("Low")
                return False
            elif guess < letter:
                print("High")
                return False
            else:
                print("unhandled exception")
                return False
        else:
            print("Invalid - more than one character")
            return False
    else:
        print("Invalid - not alphabetical")
        return False
    
def letter_game(correct_answer):
    this_guess = input("Guess the letter (provide one letter): ")
    if True == check_guess(correct_answer,this_guess):
        print("On the first try! What a mensch.")
    else:
        #second guess
        this_guess = input("Guess the letter (provide one letter): ")
        if True == check_guess(correct_answer,this_guess):
            print("On the second try! What a mensch.")
        else:
        #third guess
            this_guess = input("Guess the letter (provide one letter): ")
            if True == check_guess(correct_answer,this_guess):
                print("On the third try! Sortta slow.")
            else:
                print("You lose.")

letter_game("A")
```

### Letter Guess
**create letter_guess() function that gives user 3 guesses**
- takes a letter character argument for the answer letter
- gets user input for letter guess  
- calls check_guess() with answer and guess
- End letter_guess if 
    - check_guess() equals True, return True  
    - or after 3 failed attempts, return False

```python
# [ ] create letter_guess() function, call the function to test
def check_guess(answer,guess):
    guess = str(guess)
    if guess.isalpha():
        if len(guess) == 1:
            if guess == answer:
                print('Correct!')
                return True
            elif guess > answer:
                print('Low')
                return False
            else:
                print('High')
                return False
        else:
            print('Invalid - more than one character')
            return False
    else:
        print('Invalid - not alphabetical letter')
        return False

def letter_guess(hidden):
    user_guess = input('enter a single alphanetical character A-Z: ')
    if check_guess(hidden,user_guess) == True:
        print('Just one try to get right!')
    else:
        user_guess = input('enter a single alphanetical character A-Z: ')
        if check_guess(hidden,user_guess) == True:
            print('Second try to get right!')
        else:
           user_guess = input('enter a single alphanetical character A-Z: ') 
           if check_guess(hidden,user_guess) == True:
               print('Third try to get right!') 
           else:
               print('No more try - you lose!')

letter_guess('Q')
               
               





```

```python
# [ ] make the code above into a function rainbow_color() that has a string parameter, 
# get input and call the function and return the matching color as a string or "no match" message.
# Call the function and print the return string.
rainbow_list = {'R':'Red', 'O':'Orange','Y':'Yellow','G':'Green','B':'Blue','I':'Indigo','V':'Violet'}
def rainbow_color(your_fav):
    letter = your_fav.upper()
    if letter in rainbow_list:
       return 'Your fav color is: ' + rainbow_list[letter]
       
    else:
        return 'No match'

user_input = input('enter your fav rainbow color first letter (ROYGBIV): ')
result = rainbow_color(user_input)
print(result)


```

**Create function age_20() that adds or subtracts 20 from your age for a return value based on current age** (use `if`) 
- call the funtion with user input and then use the return value in a sentence  
example `age_20(25)` returns **5**: 
> "5 years old, 20 years difference from now"

```python
# [ ] complete age_20()
def age_20(current):
    if current >= 20:
        return current - 20
    else:
        return current + 20
    
user_input = int(input("enter ur age: "))
aba = age_20(user_input)
print(aba, "years old, 20 years from now")
```
